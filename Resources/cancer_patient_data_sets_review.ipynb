```python
# Import dependencies
import pandas as pd 
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# File to load.
cancer_patient_data_to_load = ("cancer_patient_data_sets.xlsx")

# Read the cancer patient data and store into a Panda DataFrame.
cancer_patient_data_df = pd.read_excel(cancer_patient_data_to_load)

# Display the data frame.
cancer_patient_data_df
```

```python
# Compute correlation of columns.
cancer_patient_data_df.corr()
```

```python
# Map the correlation.
fig = plt.figure(figsize = (16,6))
heatmap = sns.heatmap(cancer_patient_data_df.corr(),cmap='coolwarm',annot=True)
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16)
#plt.savefig('Heatmap.png', dpi=300, bbox_inches='tight')
```

```python
# Create a historgram of age.
fig, ax = plt.subplots(figsize=(10,6))
ax.hist(x = cancer_patient_data_df['Age'], color='blue')
#plt.savefig('Age_Histogram.png', dpi=300, bbox_inches='tight')
```

```python
# Create a dataframe for patients >= 50 years old.
cancer_patient_50_and_older = cancer_patient_data_df[cancer_patient_data_df["Age"] >= 50]
cancer_patient_50_and_older.head()
```

```python
# The number of cancer patients 50 and above
len(cancer_patient_50_and_older)
```

```python
# Create a box plot for 'level' and 'Age'
fig, ax = plt.subplots(figsize=(10,6))
sns.boxplot(data=cancer_patient_data_df, x='Level', y='Age', palette=['blue','red','yellow'])
ax.set_title('Age and Level Data', fontdict={'fontsize':18}, pad=16)
#plt.savefig('Age_and_Level_Data.png', dpi=300, bbox_inches='tight')
```

```python
# Display the correlation between level of cancer and gender.
fig, ax = plt.subplots()
sns.countplot(data=cancer_patient_data_df, x='Level', hue='Gender', palette=['blue','red'])
ax.set_title('Gender and Level Data', fontdict={'fontsize':18}, pad=16)
#plt.savefig('Gender_and_Level_Data.png', dpi=300, bbox_inches='tight')
```
