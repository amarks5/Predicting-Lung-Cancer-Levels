### Correlation betweeen the risk characteristics

# Import CSV file into PostgreSQL

```python
# Import dependencies
from sqlalchemy import create_engine
import psycopg2
from config import db_password
```

```python
def extract_transform_load():
    # Read CSV file as Pandas DataFrames.
    cancer_data_df = pd.read_csv(f'{file_dir}/cancer_patient_data_sets.csv', low_memory=False)

    # Drop unnecessary columns from DataFrame.
    cancer_data_df.drop(columns=['Patient Id'], inplace=True)
    
    #Create the connection to the PostgreSQL database.
    db_string = f"postgres://postgres:{db_password}@127.0.0.1:5432/cancer_patient_data"
    #Create the database engine.
    engine = create_engine(db_string)
    
    #Import the Cancer Data.
    cancer_data_df.to_sql(name='cancer_data', con=engine, if_exists='replace')
```
